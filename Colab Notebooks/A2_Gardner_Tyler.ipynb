Describe the graph and what it tells us. 

This boxplot describes a numeric column of our data frame based on Vehicle Odometer readings of the vehicles in our data. From this boxplot the x axis has the vehicles and the y axis represents the odometer value or reading of the vehicles. From this box plot we see that the outliers of this data range from slightly above 0 miles to over 100,000 miles. From this boxplot we can see that the Inter Quartile Range is between 60,000 and 80,000 miles which is where most vehicles odemeter readings are going to be. This box plot is a good visual to see how spread out and where your data is based off of the VehOdo column.
