# Task 4

## Code Chunk 4 - GridSearchCV

Code chunk 4 is code that uses GridSearchCV to explore all of the hyperparameter combinations from all the existing models that were created in task 2 and task 3. Cross validation is set to 3 and f1 is used as the objective best score to attempt to maximize it. After running the GridSearchCV the results and some other questions/topics are explored and explained in text blocks.

```python
# 1. using GridSearch explore ALL of the hyperparameter combination from all exiting models you have created
 # from both Step2 and Step3. Override GridSearchCV's default cv to be 3. Use f1 as your objective best score
  # to attempt to maximize.

parameters = {'learning_rate_init':[0.001,0.002,0.015],
               'hidden_layer_sizes':[(7,),(30,30),(25,10,10),(20,20),(10,10)],
              'activation':['logistic', 'tanh', 'relu'],
              'solver':['adam','sgd','lbfgs']
              }
mlp = MLPClassifier()
clf = GridSearchCV(mlp, parameters, scoring='f1', cv = 3)
clf.fit(CD_additional_modified_v2_enc, y_target)

clf.score(CD_additional_modified_v2_enc, y_target)
```

```python
# Get the data out from the GridSearchCV Code and inspect it
grid_search_df = pd.DataFrame(clf.cv_results_)

print(grid_search_df.shape)

grid_search_df.sort_values('mean_test_score',ascending=False).head() 
```

```python
# best estimator the gridsearch discovered to get the best results
clf.best_estimator_
```

1) In a text block explain how many hyperparameter combinations you will be evaluating and explain mathematically how you can be certain of this.

There are 135 hyperparameter combinations that will be evaluated. This is calculated by taking the number of learning_rate_init, hidden_layer_sizes, activation, and solver parameters and multiplying all of them. There are 3 learning rates, 5 hidden layer sizes, 3 activations, and 3 solvers. When you multiply them (3*5*3*5) you get 135.

2) In the same text block explain how many total models were evaluated when counting the cross-valildation.

Since each cross validation fold creates a model as does every hyperparameter combination you will take the value we get from the hyperparameter combinations (135) and times that by the cross validation fold size. In this case we are doing 3 fold cross validation and 135*3 equals 405 total models.

```python
# Using GridSearchCV explore at least 20 hyperparameter combinations of C and Kernel for the SVC classifier
# Use f1 as your objective best score to attempt to maximize
parameters = {'C':[5,10,15,20,30,33,40,50,60,70,80,90,100],
              'kernel':['rbf','poly','sigmoid']
              }
svc = SVC()
clf = GridSearchCV(svc, parameters,scoring='f1')
clf.fit(CD_additional_modified_v2_enc, y_target)

clf.score(CD_additional_modified_v2_enc, y_target)
```
