## Code Chunk 3 - Three Models using GridSearchCV


In code chunk three we create 3 models using GridSearcCV with the goal to have 3 good models whose hyperparameters were discovered using gridsearch. The scoring will be focused on f1 scores in these models. The first model will be a RandomForest model, the second a SVC model using sigmoid kernal, the third model will be a SVC model using radial basis kernel. The mean, std deviation and params for each of the best models will be shown.

### RandomForest Model

```python
# 1. Create 3 models using GridSearchCV. The goal here is to have 3 good models whose best hyperparameters were discovered using gridsearch.
# 2. Use f1 score for scoring 
        # 2-1. A RandomForest model (vary the n_estimators and max_depth)
        # 2-2. A SVC model using sigmoid kernel (vary the C parameter)
        # 2-3. A SVC model using radial basis kernel (again vary the C parameter)
# 3. Show the mean,std deviation, and pararms for each of the best models (ideally filter the results dataframe for rank 1)

# Random Forest Model
parameters = {'max_depth':list(range(1,15,2)),
              'n_estimators':[1,3,5,7,9,50]}
Simple_DT_Model = GridSearchCV(RandomForestClassifier(n_jobs=-1), parameters,scoring=['f1'],refit=False)
Simple_DT_Model = Simple_DT_Model.fit(CD_additional_balanced_ENC, y_target)
results = Simple_DT_Model.cv_results_
```

```python
pd.DataFrame(results)[['param_max_depth','param_n_estimators','mean_test_f1']].sort_values('mean_test_f1',ascending=False).head(10)
```

### SVC model using Sigmoid Kernel

```python
# A SVC model using sigmoid kernel (vary the C parameter)
parameters_svc_sig = {'C':list(range(1,15,2))}
SVC_SIG = GridSearchCV(SVC(kernel='sigmoid',random_state=1), parameters_svc_sig,scoring='f1',refit=True)

SVC_SIG = SVC_SIG.fit(CD_additional_balanced_ENC, y_target)
SVC_SIG_results = SVC_SIG.cv_results_
```

```python
pd.DataFrame(SVC_SIG_results).agg(['mean', np.std])
```

```python
pd.DataFrame(SVC_Vary_results).agg(['mean', np.std])
```

### SVC model using Radial Basis Kernel

```python
# A SVC model using radial basis kernel (again vary C param)
parameters_svc_radial = {'C':list(range(1,10,2))}
Model_radial = GridSearchCV(SVC(kernel='rbf', random_state=1), parameters_svc_radial, scoring='f1',refit=True)

Model_radial = Model_radial.fit(CD_additional_balanced_ENC, y_target)
radial_results = Model_radial.cv_results_
```

```python
pd.DataFrame(radial_results).agg(['mean', np.std])
```

# Task 4

## Code Chunk 4 - Voting Ensemble Classifier

In a code chunk 4 we use the best parameters found in the GridSearch for the 3 models to create/instantiate new models. These 3 models are included in a VotingClassifier object. Cross validation is then used to evaluate the VotingClassifier for f1 score. Agg is used for mean and standard deviation. A text block is then used to describe the voting ensemble method.

```python
# Using the best parameters found in the GridSearch for the 3 models to create/instantiate new models 
Model_1 = SVC(kernel='sigmoid',random_state=1)
Model_2 = RandomForestClassifier(n_estimators=50, random_state=1)
Model_3 = SVC(kernel='rbf',random_state=1)
```

```python
# Include these 3 new models in a VotingClassifier object
Ensemble_Model = VotingClassifier(
    estimators=[('svc_sig', Model_1), ('rf', Model_2), ('svc_rbf', Model_3)],
    voting='hard')
```

```python
# Using cross_validation evaluate the VotingClassifier for f1 score. Agg for mean and standard deviation
Ensemble_Model = Ensemble_Model.fit(CD_additional_balanced_ENC, y_target)
pd.DataFrame(cross_validate(Ensemble_Model,CD_additional_balanced_ENC,y_target,scoring=['f1'])).agg(['mean', np.std])
```

4 - In a text block describe the voting ensemble method.  (see the sklearn documentation)

The voting ensemble method takes all the different types and variations of models that we created and essentially combines them to make a prediction based on the results it gets from combining them all. We see this when we use SVC with sigmoid and radial base kernel along with RandomForest. When all of these are combined together, a better prediction is what we expect with the voting ensemble method.

## Code Chunk 5 - Boosting Model

In code chunk 5 we use grid search to create and evaluate a GradientBoostingClassifier for f1 score. We then filter the grid search result dataframe to show the rank 1 result of the grid search. A text block is then used to describe how a boosting model works.

```python
# Using grid search create and evaluate a GradientBoostingClassifier for f1 score.
parameters = {'max_depth':list(range(1,15,5)),
              'n_estimators':[75,150]
              }
Boost_M = GridSearchCV(GradientBoostingClassifier(), parameters,scoring=['f1'],refit=False)
Boost_M = Boost_M.fit(CD_additional_balanced_ENC, y_target)
Boost_results = Boost_M.cv_results_
```

```python
# 2. Filter the grid search result dataframe to show the rank 1 result of grid search. 
pd.DataFrame(Boost_results)[['param_n_estimators','param_max_depth','mean_test_f1']].sort_values('mean_test_f1',ascending=False).head(10)
```

3 - In a text block describe how a boosting model works. (see the sklearn documentation)

From the sklearn documentation my understanding of how boosting a model works is that it builds an additive model in a forward stage-wise fashion. In each stage regression trees are fit on the negative gradient on the binomial or multinomial deviance loss function. In normal/simplier terms it uses the negative or weak outcomes to make the prediction using a lower bias. 

# Task 6

## Code Chunk 6 - Stacked Ensemble Classifier

In code chunk 6 we use 3 models in a StackingClassifier object. For the final_estimator we use a GradientBoostingClassifier. We then use cross validation to evaluate the StackingClassifier f1 score and use Agg for mean and standard deviation. A text block is then used to describe the stacked ensemble method.

```python
# 1. Using the 3 models created in the task 3 in a StackingClassifier object
Model_1 = SVC(kernel='sigmoid',random_state=1)
Model_2 = RandomForestClassifier(n_estimators=50, random_state=1)
Model_3 = SVC(kernel='rbf',random_state=1)
```

```python
estimators = [('svc_sig', Model_1),
              ('rf', Model_2),
              ('svc_rbf', Model_3)]
```

```python
# 2. For the final_estimator use a GradientBoostingClassifier 
final_estimator = GradientBoostingClassifier(
    n_estimators=25, subsample=0.5, min_samples_leaf=25, max_features=1,
    random_state=42)
```

```python
# 3. Using cross_validation evaluate the StackingClassifierfor f1 score. Agg for mean and standard deviation
stacked_ens = StackingClassifier(
    estimators=estimators,
    final_estimator=final_estimator)
```

```python
pd.DataFrame(cross_validate(stacked_ens,CD_additional_balanced_ENC,y_target,scoring=['f1'])).agg(['mean',np.std])
```

4 - In text block describe the stacked ensemble method (see the sklearn documentation)

Stacked ensemble method is stacking the output of individual estimators and using a classifier to compute the final prediction. This allows us to use the strength of each individual estimator by using their output as input of a final estimator.
