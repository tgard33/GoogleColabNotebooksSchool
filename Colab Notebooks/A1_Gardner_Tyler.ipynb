```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
#Change the line graph color from green to red by changing facecolor='g' to facecolor='r'
#Change the alpha argument to .3 (alpha=.3)
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='r', alpha=0.3)

#Change the graph title to be "A simple line graph"
plt.title("A simple line graph")
#Add a label to the x axis as follows: plt.xlabel( "time" )
plt.xlabel( "time" )
#Add a label to the y axis as follows: plt.ylabel("amount")
plt.ylabel("amount")
plt.show()
```

#Task 3
On the above graph I changed some of the features such as the title, the facecolor, axis labels, and the alpha argument. I labeled the x axis with 'time' and the y axis with 'amount'. Labels on the axis of a graph are needed because it helps viewers understand what the graph is telling. Without axis labels its just numbers or variables and we have no idea what it is trying to show and tell us. By changing the facecolor from 'g' to 'r' the graph color that fills in the space between x axis and the line went from green to red. The alpha parameter made the red more of a lighter red on the filled in section. I changed the graph title from 'sample visualization' to 'A simple line graph'. I have also placed comments in the code of the changes made.
